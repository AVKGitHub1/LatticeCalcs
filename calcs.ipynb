```python
import numpy as np
import matplotlib.pyplot as plt
import arc
```

```python
atom = arc.Rubidium85()
pol = arc.DynamicPolarizability(atom, 5, 0, 0.5)
pol.defineBasis(5, 15)
wavelengths = np.linspace(770, 800, 1000)
pols = np.zeros_like(wavelengths)
for i, w in enumerate(wavelengths):
    ret = pol.getPolarizability(w*(10**(-9)), units='au')
    pols[i] = float(ret[0]) + 0*(3/6)*float(ret[1])
```

```python
plt.plot(wavelengths, pols)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Polarizability (au)')
plt.title('Dynamic Polarizability of Rb85')
plt.grid()
# launch matplot in interactive mode
plt.ion()
plt.show()
```
